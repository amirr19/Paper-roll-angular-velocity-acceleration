```python
import math 
```

```python
r_prime = 0.2
pi =  math.pi
start_time = 1
end_time = math.floor(pi*(r_prime**2)*1000)
```

```python
w_arr = []
t_arr = []
alpha_arr = []
tan_w = []
for t in range (start_time,end_time,1):
   w = 10 / ( math.sqrt( ( ((-10)**(-3)*t) + (pi*(r_prime)**2) ) / ( pi ) ) )
   a = (10**(-2))/(2*pi*((math.sqrt((-(10**(-3))*t+pi*(r_prime**2))/pi))**3))
   w_arr.append (w)
   t_arr.append (t)
   alpha_arr.append (a)
   tan_w.append (w/t)
```

```python
print(len(w_arr))
print(len(alpha_arr))
```

```python
# import pandas as pd
from matplotlib import pyplot as plt
data = {'w': w_arr,
        'time': t_arr,
        'alpha':alpha_arr,
        'tan-w':tan_w
       }
wplot = pd.DataFrame(data,columns=['w','time'])
wplot.plot(x ='time', y='w', kind = 'line')
plt.show()
alplot = pd.DataFrame(data,columns=['alpha','time'])
alplot.plot(x ='time', y='alpha', kind = 'line')
plt.show()
tan_w_plot = pd.DataFrame(data,columns=['tan-w','time'])
tan_w_plot.plot(x ='time', y='tan-w', kind = 'line')
plt.show()
```
